```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
from collections import Counter
from heapq import nlargest

nlp = spacy.load("en_core_web_sm")

def clean_and_extract_tags(text):
    doc = nlp(text.lower())
    tags = [token.text for token in doc if token.text.isalnum() and token.text not in STOP_WORDS]
    return " ".join(tags)

columns_to_extract_tags_from = ['category', 'brand', 'description']

for column in columns_to_extract_tags_from:
    data[column] = data[column].apply(clean_and_extract_tags)
```

```python
data['tags']=data[columns_to_extract_tags_from].apply(lambda row: ','.join(row), axis=1)
```

Rating Based Recommended **System**

```python
def recommend_top_rated_products():
    average_ratings = data.groupby(['product_name', 'reviews_count', 'brand', 'image_url'])['rating'].mean().reset_index()
    top_rated_items = average_ratings.sort_values(by='rating', ascending=False)
    top_10_products = top_rated_items.head(10)
    try:
        threshold = float(input("Enter a rating threshold between 1 and 5 (e.g., 2): "))
        if not (1 <= threshold <= 5):
            print("Please enter a value between 1 and 5.")
            return
    except ValueError:
        print("Invalid input. Please enter a numeric value.")
        return
    filtered_products = top_10_products[top_10_products['rating'] >= threshold]
    top_5_products = filtered_products.head(5)
    print("\nRating-Based Recommendation System: (Trending Products)")
    return top_5_products[['product_name', 'rating', 'reviews_count', 'brand', 'image_url']]

recommended_products = recommend_top_rated_products()
if recommended_products is not None:
    redata = pd.DataFrame(recommended_products)
    redata.drop(columns='image_url', inplace=True)
    print(redata)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def recommend_similar_items():
    tfidf_vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix_content = tfidf_vectorizer.fit_transform(data['tags'])
    cosine_similarities_content = cosine_similarity(tfidf_matrix_content, tfidf_matrix_content)

    try:
        item_name = input("Enter a product name or keyword (e.g., 'nail polish'): ")
        item_index = data[data['product_name'].str.contains(item_name, case=False, na=False)].index[0]
    except IndexError:
        print("No matching product found. Please try again with a valid product name or keyword.")
        return

    similar_items = list(enumerate(cosine_similarities_content[item_index]))
    similar_items = sorted(similar_items, key=lambda x: x[1], reverse=True)
    top_similar_items = similar_items[1:11]
    recommended_items_indices = [x[0] for x in top_similar_items]

    recommendations = data.iloc[recommended_items_indices][['product_name', 'reviews_count', 'brand']]
    print("\nTop 10 Recommended Products:")
    return recommendations

recommended_products = recommend_similar_items()
if recommended_products is not None:
    print(recommended_products)
```
